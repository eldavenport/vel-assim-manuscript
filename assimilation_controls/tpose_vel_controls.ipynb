```python
import matplotlib.pyplot as plt
import cmocean.cm as cmo
from mitgcm_assim.ctrls import load_controls_and_sensitivities

sep_data_dir = '/data/SO3/edavenport/tpose6/sep2012/velocity_assim/run_ad'
nov_data_dir = '/data/SO3/edavenport/tpose6/nov2012/run_iter20'
jan_data_dir = '/data/SO3/edavenport/tpose6/jan2013/run_iter14'
mar_data_dir = '/data/SO3/edavenport/tpose6/mar2013/run_iter16'

grid_dir = '/data/SO6/TPOSE_diags/tpose6/grid_6/'

sep_final_iter = 22
nov_final_iter = 20
jan_final_iter = 14
mar_final_iter = 16 

folder_name = 'assimilation_controls/'

# Load all control variables and sensitivities into a Dataset
# This automatically loads grid information and applies proper MITgcm coordinate names
ds_sep = load_controls_and_sensitivities(sep_data_dir, grid_dir, sep_final_iter)
ds_nov = load_controls_and_sensitivities(nov_data_dir, grid_dir, nov_final_iter)
ds_jan = load_controls_and_sensitivities(jan_data_dir, grid_dir, jan_final_iter)
ds_mar = load_controls_and_sensitivities(mar_data_dir, grid_dir, mar_final_iter)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import cmocean
import cartopy.crs as ccrs
import cartopy.feature as cfeature

datasets = {'Sep': ds_sep, 'Nov': ds_nov, 'Jan': ds_jan, 'Mar': ds_mar}

variables = {
    'xx_atemp':  ('Atmos. Temp.', cmocean.cm.amp),
    'xx_precip': ('Precip.',      cmocean.cm.rain),
    'xx_uwind':  ('U Wind',       cmocean.cm.speed),
    'xx_vwind':  ('V Wind',       cmocean.cm.speed),
    'xx_swdown':  ('Shortwave Rad.', cmocean.cm.solar),
}

# Compute RMS maps for all variables and datasets
rms_maps = {
    var: {label: np.sqrt((ds[var]**2).mean(dim='time')) for label, ds in datasets.items()}
    for var in variables
}

# Compute shared colorbar limits per row (variable)
vlims = {
    var: (
        min(rms.min().values for rms in rms_maps[var].values()),
        max(rms.max().values for rms in rms_maps[var].values())
    )
    for var in variables
}

# TAO/TRITON moorings in 0-360 longitude
tao_moorings = [
    (220, 0), (190, 0), (250, 0),   # highlighted: 140W, 170W, 110W
    (265, 0), (235, 0), (205, 0),
    (220, 2), (220, -2), (220, 5), (220, -5), (220, 8), (220, -8),
    (190, 2), (190, -2), (190, 5), (190, -5), (190, 8), (190, -8),
    (250, 2), (250, -2), (250, 5), (250, -5), (250, 8), (250, -8),
]
highlight = {(220, 0), (190, 0), (250, 0)}

n_rows = len(variables)
n_cols = len(datasets)
proj = ccrs.PlateCarree(central_longitude=180)  # centers on Pacific for 0-360 data

fig = plt.figure(figsize=(20, 15))
gs = gridspec.GridSpec(n_rows, n_cols + 1, width_ratios=[1, 1, 1, 1, 0.03], wspace=0.15, hspace=0.12)

all_axes = np.empty((n_rows, n_cols), dtype=object)

for row, (var, (title, cmap)) in enumerate(variables.items()):
    vmin, vmax = vlims[var]
    cax = fig.add_subplot(gs[row, n_cols])

    for col, (label, rms) in enumerate(rms_maps[var].items()):
        ax = fig.add_subplot(gs[row, col], projection=proj)
        ax.set_aspect('auto')  # <-- add this line
        all_axes[row, col] = ax

        # Land
        ax.add_feature(cfeature.COASTLINE, linewidth=0.8, edgecolor='black')
        ax.add_feature(cfeature.LAND, facecolor='lightgray', zorder=1)

        # Extent in 0-360
        xmin = float(rms.XC.min())
        xmax = float(rms.XC.max())
        ax.set_extent([xmin, xmax, -17, 17], crs=ccrs.PlateCarree())

        im = rms.plot(ax=ax, add_colorbar=False, cmap=cmap, vmin=vmin, vmax=vmax,
                      transform=ccrs.PlateCarree(), x='XC', y='YC')

        # Mooring overlays
        for (lon, lat) in tao_moorings:
            color = 'red' if (lon, lat) in highlight else 'white'
            ax.plot(lon, lat, marker='o', color=color, markersize=4,
                    markeredgecolor='black', markeredgewidth=0.5,
                    transform=ccrs.PlateCarree(), zorder=5)

        # Gridlines and tick labels
        gl = ax.gridlines(draw_labels=True, linewidth=0.3, color='gray',
                          alpha=0.5, linestyle='--', crs=ccrs.PlateCarree())

        # Only show left labels on left column
        gl.left_labels = (col == 0)
        gl.right_labels = False
        gl.top_labels = False
        # Only show bottom labels on bottom row
        gl.bottom_labels = (row == n_rows - 1)

        gl.xlabel_style = {'size': 12}
        gl.ylabel_style = {'size': 12}

        if row == 0:
            ax.set_title(label, fontsize=14, fontweight='bold')
        else:
            ax.set_title('')

        if col == 0:
            ax.text(0.02, 0.97, title, transform=ax.transAxes,
                    fontsize=12, fontweight='bold', va='top', ha='left',
                    bbox=dict(boxstyle='round,pad=0.3', facecolor='white', 
                              alpha=1.0, edgecolor='gray', linewidth=0.5))

    fig.colorbar(im, cax=cax, label='RMS (units of $\\sigma$)')

plt.savefig('assimilation_controls/rms_atmos_controls.png', bbox_inches='tight')
plt.show()
```
