# GLORYS Reanalysis Download

Downloads GLORYS12 daily reanalysis (SSH, U, V) for the equatorial Pacific:

| Parameter | Value |
|-----------|-------|
| Longitude | 180–260 °E |
| Latitude  | 10 °S – 10 °N |
| Depth     | 0–500 m |
| Variables | SSH, U, V |
| Period    | 01 Sep 2012 – 30 Jun 2013 |

Files are saved monthly to `glorys_data/`.

**Credentials**: run `copernicusmarine login` once in the terminal, or set the
environment variables `COPERNICUSMARINE_SERVICE_USERNAME` /
`COPERNICUSMARINE_SERVICE_PASSWORD` before launching this notebook.

```python
import sys
from pathlib import Path

# Add forecasts/ to path so glorys_download is importable from anywhere
sys.path.insert(0, str(Path("__file__").resolve().parent))

from forecasts.glorys_download import download_glorys
```

```python
OUTPUT_DIR = "glorys_data"

files = download_glorys(
    lon_min=180,
    lon_max=260,
    lat_min=-10,
    lat_max=10,
    depth_min=0,
    depth_max=-500,
    variables=["SSH", "U", "V"],
    time_start="09-01-2012",
    time_end="06-30-2013",
    output_dir=OUTPUT_DIR,
)
```

```python
print(f"Downloaded {len(files)} file(s):")
for f in files:
    size_mb = Path(f).stat().st_size / 1e6 if Path(f).exists() else 0
    print(f"  {f.name}  ({size_mb:.1f} MB)")
```

```python
import xarray as xr

# Quick sanity check on the first file
if files:
    ds = xr.open_dataset(files[0])
    print(ds)
    ds.close()
```
