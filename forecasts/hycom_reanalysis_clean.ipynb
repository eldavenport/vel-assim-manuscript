# HYCOM Reanalysis Download

Downloads HYCOM GLBv0.08/expt_53.X daily reanalysis (SSH, U, V) for the equatorial Pacific:

| Parameter | Value |
|-----------|-------|
| Longitude | 180–260 °E |
| Latitude  | 10 °S – 10 °N |
| Depth     | 0–500 m |
| Variables | SSH, U, V |
| Period    | 01 Sep 2012 – 30 Jun 2013 |

Files are saved monthly to `hycom_data/`.  No authentication is required.

**Test cell** below downloads one week (< 1 GB) to verify connectivity before
running the full download.

```python
import sys
from pathlib import Path

# Add forecasts/ to path so hycom_download is importable from anywhere
sys.path.insert(0, str(Path("__file__").resolve().parent))

from forecasts.hycom_download import download_hycom
```

```python
# ── Test download: 1 week, same region ──────────────────────────────────────
# Estimated size: ~200 MB (well under 1 GB)
test_files = download_hycom(
    lon_min=180,
    lon_max=260,
    lat_min=-10,
    lat_max=10,
    depth_min=0,
    depth_max=-500,
    variables=["SSH", "U", "V"],
    time_start="09-01-2012",
    time_end="09-07-2012",
    output_dir="/data/SO3/edavenport/hycom_data",
)
```

```python
import xarray as xr

# Sanity check on test file
if test_files:
    ds = xr.open_dataset(test_files[0])
    print(ds)
    ds.close()
```

```python
# ── Full download: Sep 2012 – Jun 2013 ──────────────────────────────────────
OUTPUT_DIR = "hycom_data"

files = download_hycom(
    lon_min=180,
    lon_max=260,
    lat_min=-10,
    lat_max=10,
    depth_min=0,
    depth_max=-500,
    variables=["SSH", "U", "V"],
    time_start="09-01-2012",
    time_end="06-30-2013",
    output_dir=OUTPUT_DIR,
)
```

```python
print(f"Downloaded {len(files)} file(s):")
for f in files:
    size_mb = Path(f).stat().st_size / 1e6 if Path(f).exists() else 0
    print(f"  {f.name}  ({size_mb:.1f} MB)")
```
